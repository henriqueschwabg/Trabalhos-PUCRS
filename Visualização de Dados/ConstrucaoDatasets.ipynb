```
dfs = [df2008, df2009, df2010, df2011, df2012, df2013, df2014, df2015, df2016, df2017, df2018]

# inserindo em cada dataframe uma coluna com UF da ocorrência
for df in dfs:
    df['UFOCOR'] = df['CODMUNOCOR'].apply(lambda x: int(str(x)[:2]))

# inserindo em cada dataframe uma coluna com UF de residência
for df in dfs:
    df['UFRES'] = df['CODMUNRES'].apply(lambda x: int(str(x)[:2]))

df2008.head()
```

```
# inserindo em cada dataframe uma coluna com categoria do CID10 (causas)

for df in dfs:
    df['CATCID10'] = df['CAUSABAS'].apply(lambda x: x[:3])

df2008.head()
```

```
# Q4 >> Criando df para homicidios por raça/cor e estado por ano

lista_homicidios_racacor = []

for cod in cod_estados:
    array = []  
    for df in dfs:
        for cor in [1, 2, 3, 4, 5]:
            array.append((((df['CIRCOBITO']==3) & (df['UFOCOR']==cod)) & (df['RACACOR']==cor)).sum())
    lista_homicidios_racacor.append(array)

colunas_racacor = []
for ano in anos:
    for cor in ['Branca', 'Preta', 'Amarela', 'Parda', 'Indígena']:
        colunas_racacor.append(ano+cor)

df_homicidios_racacor = pd.DataFrame(lista_homicidios_racacor, columns=colunas_racacor)

df_homicidios_racacor.insert(loc=0, column='Cód', value=df_mortes_ufocor['Cód'].values)
df_homicidios_racacor.insert(loc=1, column='UF', value=df_mortes_ufocor['Unidade da Federação'].values)

df_homicidios_racacor['lat'] = df_latlon_estados['latitude']
df_homicidios_racacor['lon'] = df_latlon_estados['longitude']

df_homicidios_racacor.head()
```

```
# Q4 >> Criando df para homicidios por sexo e estado por ano

# Tem o valor 0 em alguns dataframes, que significa 'ignorado', mas em outros tem o valor 9 (e não tem o 0). Vou considerar 9 como ignorado também.
# Os valores "ignorados" representam cerca de 0,1% nos dataframes e serão juntados ao valor 'masculino', que são a grande maioria

lista_homicidios_sexo = []

for cod in cod_estados:
    array = []  
    for df in dfs:
        array.append(((df['UFOCOR']==cod) & ((df['CIRCOBITO'] == 3) & ((df['SEXO'] == 1) | ((df['SEXO'] == 0) | (df['SEXO'] == 9))))).sum())
        array.append(((df['UFOCOR']==cod) & ((df['CIRCOBITO'] == 3) & (df['SEXO'] == 2))).sum())
    lista_homicidios_sexo.append(array)

print(lista_homicidios_sexo)

colunas_sexo = []
for ano in anos:
    for sexo in ['Masculino', 'Feminino']:
        colunas_sexo.append(ano+sexo)

df_homicidios_sexo = pd.DataFrame(lista_homicidios_sexo, columns=colunas_sexo)

df_homicidios_sexo.insert(loc=0, column='Cód', value=df_mortes_ufocor['Cód'].values)
df_homicidios_sexo.insert(loc=1, column='UF', value=df_mortes_ufocor['Unidade da Federação'].values)

df_homicidios_sexo['lat'] = df_latlon_estados['latitude']
df_homicidios_sexo['lon'] = df_latlon_estados['longitude']

df_homicidios_sexo.head()
```

```
# Q5 >> Criando dataframe com as 10 causas mais frequentes de mortes violentas (onde CIRCOBITO é 1-acidente ou 2-suicidio ou 3-homicidio ou 4-outros)
# O valor 9-ignorado não será utilizado para a contagem

array_cod_estados = []
array_ano = []
array_ordem = []
array_codigos = []
array_valores = []

for cod in cod_estados:
    for df, ano in zip(dfs, anos):
        for i in range(10):
            array_cod_estados.append(cod)
            array_ano.append(ano)
            array_ordem.append(i+1)
        for index in df['CATCID10'][(df['UFOCOR']==cod) & (((df['CIRCOBITO']==1) | (df['CIRCOBITO']==2)) | ((df['CIRCOBITO']==3) | (df['CIRCOBITO']==4)))].value_counts()[:10].index:
            array_codigos.append(index)
        for value in df['CATCID10'][(df['UFOCOR']==cod) & (((df['CIRCOBITO']==1) | (df['CIRCOBITO']==2)) | ((df['CIRCOBITO']==3) | (df['CIRCOBITO']==4)))].value_counts()[:10].values:
            array_valores.append(value)
```

```
lista_causas_violentas = []

lista_causas_violentas.append(array_cod_estados)
lista_causas_violentas.append(array_ano)
lista_causas_violentas.append(array_ordem)
lista_causas_violentas.append(array_codigos)
lista_causas_violentas.append(array_valores)

df_causas_violentas = pd.DataFrame(lista_causas_violentas).transpose()

df_causas_violentas.columns = ['cod_uf', 'ano', 'ordem', 'cod_causa', 'quantidade']

df_causas_violentas
```
