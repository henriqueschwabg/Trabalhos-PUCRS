```python
%matplotlib inline
import numpy as np
from matplotlib import pyplot as plt
import csv
```

## Questão 1 (1.0 pts)

Leia o arquivo `heart.csv` e imprima os seguintes dados sobre cada atributo (exceto o atributo alvo/target):
- Valor mínimo e máximo
- Valor médio
- Desvio padrão
- Mediana

```python
########################
# Inicio do seu codigo #
########################

# estrutura para guardar os dados
dataset = {
    'age': [],
    'sex': [],
    'cp': [],
    'trestbps': [],
    'chol': [],
    'fbs': [],
    'restecg': [],
    'thalach': [],
    'exang': [],
    'oldpeak': [],
    'slope': [],
    'ca': [],
    'thal': [],
    'target': []
}

# lendo o arquivo csv e preenchendo o array
with open('heart.csv', 'r') as f:
    csv_file = csv.reader(f, delimiter=',', quotechar='"')
    for i, row in enumerate(csv_file):
        if i == 0:
            continue
        dataset['age'].append(int(row[0]))
        dataset['sex'].append(int(row[1]))
        dataset['cp'].append(int(row[2]))
        dataset['trestbps'].append(int(row[3]))
        dataset['chol'].append(int(row[4]))
        dataset['fbs'].append(int(row[5]))
        dataset['restecg'].append(int(row[6]))
        dataset['thalach'].append(int(row[7]))
        dataset['exang'].append(int(row[8]))
        dataset['oldpeak'].append(float(row[9]))
        dataset['slope'].append(int(row[10]))
        dataset['ca'].append(int(row[11]))
        dataset['thal'].append(int(row[12]))
        dataset['target'].append(int(row[13]))

# transformando as listas em arrays numpy
for k in dataset.keys():
    dataset[k] = np.array(dataset[k])    
    
# obtendo as informações estatísticas
for i, d in enumerate(dataset):
    if i == 13:
        break
    print(f'{d} \nValor médio: {np.mean(dataset[d])}\
            \nValor mínimo: {np.min(dataset[d])}\
            \nValor máximo: {np.max(dataset[d])}\
            \nDesvio-padrão: {np.std(dataset[d])}\
            \nMediana: {np.median(dataset[d])}\n')

########################
#   Fim do seu codigo  #
########################
```

## Questão 3 (3.0 pts)

Realize as seguintes operações sobre os atributos idade (`age`) e colesterol (`chol`):
1. Plotar um gráfico de dispersão com o atributo colesterol no eixo X e a idade no eixo Y. Utilize o atributo target para colorir os pontos de acordo com a sua classe (escolha duas cores diferentes).
2. Normalize os atributos idade (`age`) e colesterol (`chol`) utilizando a estratégia `MinMax`. A normalização MinMax é dada pela fórmula abaixo:
$$x^{(i)}_{norm}=\frac{x^{(i)} - x^{(i)}_{min}}{x^{(i)}_{max} - x^{(i)}_{min}}$$

Onde $x^{(i)}_{norm}$ é o atributo $x^{(i)}$ do conjunto de dados normalizado.

3. Após normalizar os atributos, faça a plotagem do passo 1 novamente, agora com os atributos normalizados.

```python
########################
# Inicio do seu codigo #
########################

colors = np.array(["black", "green"])

# plotando um gráfico de dispersão
fig, ax = plt.subplots()
ax.scatter(dataset['chol'], dataset['age'], c=colors[dataset['target']])
ax.set_xlabel('Colesterol')
ax.set_ylabel('Idade')
plt.show()

# realizando a normalização das variáveis 'age' e 'chol'
norm_age = (dataset['age'] - np.min(dataset['age'])) / (np.max(dataset['age']) - np.min(dataset['age']))
norm_chol = (dataset['chol'] - np.min(dataset['chol'])) / (np.max(dataset['chol']) - np.min(dataset['chol']))

# plotando um gráfico de dispersão com as variáveis normalizadas
fig, ax = plt.subplots()
ax.scatter(norm_chol, norm_age, c=colors[dataset['target']]) 
ax.set_xlabel('Colesterol')
ax.set_ylabel('Idade')
plt.show()
########################
#   Fim do seu codigo  #
########################
```

## Questão 5 (2.5 pts)

Faça dois graficos que pizza (*pie chart*) que contenham informações sobre os tipos de dores no peito (`cp`) com o tipo de dor (0, 1, 2, 3) e a frequência. Faça um plot para homens e outro para mulheres.

```python
########################
# Inicio do seu codigo #
########################

masc, nmro_fem = np.unique(dataset['cp'][dataset['sex'] == 0], return_counts = True)
fem, nmro_masc = np.unique(dataset['cp'][dataset['sex'] == 1], return_counts = True)

labels = ['Tipo 0', 'Tipo 1', 'Tipo 2', 'Tipo 3']

def func(pct, allvals):
    absolute = int(pct/100.*np.sum(allvals))
    return "{:.2f}%".format(pct, absolute)

fig, axes = plt.subplots(nrows=1, ncols=2)

axes[0].set_title('Homens')
wedges, texts, autotexts = axes[0].pie(nmro_masc, autopct=lambda pct: func(pct, nmro_masc),
                                  textprops=dict(color="white"))

axes[1].set_title('Mulheres')
wedges, texts, autotexts = axes[1].pie(nmro_fem, autopct=lambda pct: func(pct, nmro_fem),
                                  textprops=dict(color="white"))

axes[1].legend(wedges, labels,
          title="Tipos de Dor",
          loc="center left",
          bbox_to_anchor=(1, 0, 0.5, 1))

plt.show()

########################
#   Fim do seu codigo  #
########################
```
