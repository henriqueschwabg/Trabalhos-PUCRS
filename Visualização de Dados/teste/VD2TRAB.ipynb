```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
rc = {'figure.figsize':(12,8),
      'axes.facecolor':'white',
      'axes.grid' : True,
      'grid.color': '.8',
      'font.family':'Times New Roman',
      'font.size' : 15}
plt.rcParams.update(rc)
```

Para esta solução resolvi destacar que, com a saída do antigo colaborador, os números de atendimentos que foram registrados como 'atendido' diminuíram expressivamente em comparação aos meses anteriores, quando o colaborador ainda não havia se aposentado. A chegada do novo colaborador não foi o suficiente para manter o mesmo padrão, nem mesmo com o passar dos meses.
O gráfico de barras abaixo mostra os números de todos os meses do ano.

```python
N = 12

ind = np.arange(N)
width = 0.45

ig, ax = plt.subplots(figsize=(16, 10))

atendido = (180, 194, 261, 169, 200, 171, 152, 202, 180, 149, 159, 177)
recebido = (180, 194, 250, 169, 181, 160, 143, 156, 136, 104, 114, 110)

rects1 = ax.bar(ind, atendido, width, color='r')
rects2 = ax.bar(ind + width, recebido, width, color='y')

ax.set_title('Suporte Técnico - Métricas de Atendimento (2020)')
ax.set_ylabel('atendimentos')
ax.set_xlabel('mês')
ax.set_xticks(ind + width / 2)
ax.set_xticklabels(('Jan', 'Fev', 'Mar', 'Abr', 'Mai', 'Jun', 'Jul', 'Ago', 'Set', 'Out', 'Nov', 'Dez'))

ax.legend((rects1[0], rects2[0]), ('Recebido', 'Atendido'))


def autolabel(rects):
    for rect in rects:
        height = rect.get_height()
        ax.text(rect.get_x() + rect.get_width()/2., 1.*height,
                '%d' % int(height),
                ha='center', va='bottom')

autolabel(rects1)
autolabel(rects2)
```
